# Linear Regression

Objectives: linear regression equation, loss, meshgrid, contour plot and 3d plot

# Supervised Learning

The purpose of this notebook is to explore the linear regression model.



```python
# Math library
import numpy as np
# Plotting library
import matplotlib.pyplot as plt
```

```python
# Create some input / output data
# x represents input data / y represent output data
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

print(x)
print(y)
```

###1D Linear Regression Equation:

<center>

![picture](https://drive.google.com/uc?export=view&id=1LoOuhxpoYk5mhNN5P57xPZl97Ntiel0a)

```python
# Define 1D linear regression model
def f(x, phi0, phi1):

  y = phi0 + phi1*x


  return y
```

```python
# Function to help plot the data
def plot(x, y, phi0, phi1):
    fig,ax = plt.subplots()
    ax.scatter(x,y)
    plt.xlim([0,2.0])
    plt.ylim([0,2.0])
    ax.set_xlabel('Input, $x$')
    ax.set_ylabel('Output, $y$')
    # Draw line
    x_line = np.arange(0,2,0.01)
    y_line = f(x_line, phi0, phi1)
    plt.plot(x_line, y_line,'b-',lw=2)

    plt.show()
```

This figure will be used for below given cells.

```python
# Set the intercept and slope as in graph b in figure given above
phi0 = 0.4 ; phi1 = 0.8
# Plot the data and the model
plot(x,y,phi0,phi1)
```

###Loss Equation:

<center>

![picture](https://drive.google.com/uc?export=view&id=1c3-37pkURM1nHaG-DHFCVg4wwuL7f8rN)

</center>

```python
# Function to calculate the loss
def compute_loss(x,y,phi0,phi1):

  loss = np.sum((f(x,phi0,phi1)-y)**2)
  return loss
```

```python
# Compute the loss for our current model
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =7.07')
```

```python
# Set the intercept and slope as in graph c in figure given above
phi0 = 1.0 ; phi1 =0.3
# Plot the data and the model
plot(x,y,phi0,phi1)
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =10.28')
```

#Understanding MeshGrid

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define two coordinate vectors (1D arrays)
x = np.linspace(-3, 3, 5)  # 5 points between -3 and 3
y = np.linspace(-2, 2, 4)  # 4 points between -2 and 2

# 2. Use meshgrid to create a 2D grid from the 1D coordinate vectors
X, Y = np.meshgrid(x, y)

# 3. Display the grids
print("X grid:")
print(X)
print("\nY grid:")
print(Y)

# 4. Visualize the grid points
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='red', marker='o')  # Plot points on the grid
plt.title("2D Grid Generated by meshgrid")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.show()
```

# plotting Contour

```python
import numpy as np
import plotly.graph_objects as go

# Define the function to plot
def plot_function(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

# Generate the grid for the plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = plot_function(X, Y)

# Create the 3D surface plot
surface = go.Surface(z=Z, x=X, y=Y, colorscale='Magma')

# Create the 2D contour plot
contour = go.Contour(z=Z, x=x, y=y, colorscale='Magma')

# Set up the layout with two subplots
fig = go.Figure(data=[surface, contour])

# Update layout for better visualization
fig.update_layout(
    title="Interactive 3D Surface Plot and 2D Contour Plot",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    scene_aspectmode='cube',
    margin=dict(l=0, r=0, b=0, t=40)
)

# Show the figure
fig.show()
```
